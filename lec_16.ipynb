# Important Function Dictionary

```python
# update()
#         --> ll itemes preent in dictionary 1 will be added in dictionary 2.
```

```python
# x's all element update in d

x={1:"Arman",2:"Aryan",3:"Lucky"}
d={4:"Dhairya",5:"xyz"}
d.update(x)
print(d)
print(x)
```

```python
# d's all element update in x

x={1:"Arman",2:"Aryan",3:"Lucky"}
d={4:"Dhairya",5:"xyz"}
x.update(d)
print(d)
print(x)
```

```python
# get()
#      --> To get the value associated with key.
#      --> d.get(key)
#      --> d.get(key,defualtvalue)

d={10:"Lucky",20:"Aryan",30:"Arman"}
print(d[10])
print(d.get(10))
print(d.get(10,"Guest")) # if 10 key is available in dictionary then it will gave iyt's value
print(d.get(40,"Guest")) # otherwise it will gave defualtvalue.
print(d.get(40))
print(d[40])
```

```python
# setdefault():
#              --> d.setdefault(k,v)
#              --> if the key is alredy available then this function returns corspnding value.
#              --> if key is not available then specified key-value will be added as new item to the dictionary.

d={10:"Lucky",20:"Aryan",30:"Arman"}
print(d.setdefault(40,"Pavan"))
print(d)
print(d.setdefault(10,"Sachin"))
```

# WAP with user define function that reads word from the paragraph and stores them as a keys in dictionary and the count frequency of as a values.

```python
s="dog the quick brown the jumps over the lazy dog."
def dicti(s):
    d={}
    sp=s.split(" ")
    c=0
    for i in sp:
        c=sp.count(i)
        d.setdefault(i,c)
    print(d)
dicti(s)
```

# WAP to create a dictionary with key as the first charactor and values as a list of words starting with thier charactor.

```python
d="Don't wait for your feelings to change to tacke action"
sp=d.split(" ")
r={}
for i in sp:
    k=i[0]
    if k in r:
        r[k].append(i)
    else:
        r[k]=[i]
print(r)
```

# Set

```python
# --> Group of uniqe value as a single entity.
# --> Duplicates not allowed.
# --> Insertion order not presented but we can sort element.
# --> Indexing and slicing are not allowed for set.
# --> Heterogeneous objects are allowed.
# --> we can represent set elements within curly bressis with comma seperation.
# --> we can apply mathemetical operations like union,intersection,difference,sementic,difference.

s=set()
print(s)
print(type(s))
```

```python
s=set()
s={10,5,15,6,28,2,6,8,45}
print(s)
s={10,5,15,5,6,5,28}
print(s)
print(type(s))
```

```python
s=set([1,1,3,3,2,5,7,9,21])
print(s)
s=set(range(5))
print(s)
```

```python
s={"Apple","Banana","Chery","Civi"}
print(s)
```

# Mathematical Operation

```python
# 1. union():
#            --> union(x) - return all element present in both sets.
#            --> x|y

x={10,20,30,40}
y={30,40,50,60,70,80}
print(x.union(y))
print(x)
```

```python
x={10,20,30,40}
y={30,40,50,60,70,80}
print(y.union(x))
print(y)
```

```python
# 2. intersection :
#                  --> returns comman element present in both set.
#                  --> x.intersection(y)
#                  --> x&y

x={10,20,30,40}
y={30,40,50,60,70,80}
print(x.intersection(y))
print(x)
```

```python
x={10,20,30,40}
y={30,40,50,60,70,80}
print(y.intersection(x))
print(y)
```

```python
# 3. difference() :
#                  --> retuns elements present in x but not y.
#                  --> x.difference(y)
#                  --> x-y

x={10,20,30,40}
y={30,40,50,60,70,80}
print(x.difference(y))
print(x)
```

```python
x={10,20,30,40}
y={30,40,50,60,70,80}
print(y.difference(x))
print(y)
```

```python
# 4. symmetric_difference() :
#                            --> returns element present in either x or y but not in both.

x={10,20,30,40}
y={30,40,50,60,70,80}
print(x.symmetric_difference(y))
print(x)
```

```python
x={10,20,30,40}
y={30,40,50,60,70,80}
print(y.symmetric_difference(x))
print(y)
```

```python
# 5. issubset()

x={10,20,30,40}
y={10,20,30,40,50,60,70}
z=x.issubset(y)
p=y.issubset(x)
print(z)
print(p)
```

```python
# 6. issuperset()

x={10,20,30,40}
y={10,20,30,40,50,60,70}
z=x.issuperset(y)
p=y.issuperset(x)
print(z)
print(p)
```
