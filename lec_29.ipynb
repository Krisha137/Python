```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr3=np.array([7,8,9])
arr=np.concatenate((arr1,arr2,arr3))
print(arr)
```

```python
arr1=np.array([[1,2],[5,6]])
arr2=np.array([[3,4],[7,8]])
arr=np.concatenate((arr1,arr2),axis=0) # bydefult axis is 0.
print(arr)
```

# WAP to make number gasing name.attempt is 7

```python
from numpy import random
x=random.randint(100)
i=0
while(i<7):
    print("-"*50)
    print("You have ",7-i," Attempt.")
    n=int(input("Enter Number :"))
    if n==x:
        print("🎉🎉------------Congratulation----------🎉🎉")
        print("You Win😎")
        print("You guess correct number.")
        break
    if n>x:
        print("Number is Lower then your Number.")
    else:
        print("Number is Greater then your Number.")
    i+=1
    print("-"*50)
else:
    print("----game over----")
    print("You can not able to guess number.😌")
    print("Correct Number is :",x)
```
