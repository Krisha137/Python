# Pandas Series

```python
# --> pandas series is like a column in a table 
# --> it is a 1-D array holding data of any type

import pandas as pd
a=[1,7,2]
myvar=pd.Series(a)
print(myvar)
print(type(myvar))
print(myvar[1])
```

```python
# If you want index according to you then---

import pandas as pd
a=[1,7,2]
myvar=pd.Series(a,index=["x","y","z"])
print(myvar)
print(type(myvar))
print(myvar[1]) # valid in new version
print(myvar["y"])
```
