```python
import requests
from bs4 import BeautifulSoup
url="https://www.politifact.com/factchecks/"
result = requests.get(url)
content = result.text
soup=BeautifulSoup(content,"html.parser")
# print(soup.prettify())
```

```python
import pandas as pd
li=[] 
Line=soup.find_all("div",class_="m-statement__quote")
for i in Line:
    li.append(i.find("a",href=True).text.strip())
 # li
```

```python
so=[]
box=soup.find_all("div",class_="m-statement__meta")
for i in box:
    so.append(i.find("a",href=True).text.strip())
# so
```

```python
date=[]
box=soup.find_all("div",class_="m-statement__content")
for i in box:
    a=i.find("footer").text.strip().split("•")
    date.append(a[1])
# date
```

```python
link="hrrps://politifact.com"
lin=[]
box=soup.find_all("div",class_="m-statement__quote")
for i in box:
     a=link+i.find("a")["href"].strip()
     lin.append(a)
# lin
```

```python
df=pd.DataFrame({"Statement":li,
   "Source":so,
   "Date":date,
   "Link":lin})
df
```

# For New Pages

```python
from bs4 import BeautifulSoup
import requests
upperframe=[]
for page in range(1,5):
    url="https://www.politifact.com/factchecks/?page="+str(page)
    result = requests.get(url)
    content = result.text
    soup=BeautifulSoup(content,"html.parser")
    frame=[]
    links=soup.find_all("li",class_="o-listicle__item")
    for j in links:
        statement=j.find("div",class_="m-statement__quote").text.strip()
        link="hrrps://politifact.com"
        link+=j.find("div",class_="m-statement__quote").find("a")["href"].strip()
        date=j.find("div",class_="m-statement__body").find("footer").text.split("•")[-1]
        source=j.find("a",class_="m-statement__name").text.strip()
        frame.append((statement,link,date,source))
    upperframe.extend(frame)
print(upperframe)
```

```python
# Geocoding API + 5 day weather forecast

import requests,json
import pandas as pd

def make_dataframe():
    api_key=input("Enter API Key :")
    city=input("Enter City Name :")
    requests_url1=f'http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}'

    # requests_url="http://api.openweathermap.org/geo/1.0/direct?q="+city+"&appid="+api_key
    response1=requests.get(requests_url1)
    coordinates=response1.json()
    # print(coordinates)
    # print(json.dumps(coordinates,indent=5))
    print("-"*50)
    lat=coordinates[0]["lat"]
    lon=coordinates[0]["lon"]
    requests_url4=f'https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}'
    response4=requests.get(requests_url4)
    coordinates=response4.json()
    date_time=[]
    desc=[]
    temp=[]
    pre=[]
    humi=[]
    visi=[]
    wind=[]
   # print(coordinates)
    for i in coordinates["list"]:
        date_time.append(i["dt_txt"])
        desc.append(i["weather"][0]["description"])
        temp.append(i["main"]["temp"])
        pre.append(i["main"]["pressure"])
        humi.append(i["main"]["humidity"])
        visi.append(i["visibility"])
        wind.append(i["wind"]["speed"])
    df=pd.DataFrame({"Date_Time":date_time,"Description":desc,"Tempreture":temp,"Pressure":pre,"Humidity":humi,"Visibility":visi,"Wind_Speed":wind})
    # print(df)
    return df
df=make_dataframe()
df
```

```python
# You can accses multiple pages content in one time
import requests
url="https://jsonplaceholder.typicode.com/posts"
page=1
while page<=3:
    params={
    "_page":page,
    "_limit":5
    }
    response=requests.get(url,params=params)
    data=response.json()
    for i in data:
        print(i["id"],i["title"])
    page+=1
```
