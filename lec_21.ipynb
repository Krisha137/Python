# WAP to count the frequency of words in a given text file.

```python
f=input("Enter file name :")
fn=f+".txt"
n=open(fn)
d=n.read()
s=d.split()
di=dict()
for i in s:
    if i in di:
        di[i]+=1
    else:
        di[i]=1
n.close()
for i in di:
    print(i," - ",di[i])
```

# WAP to accept string or sentence from the user till the user enter END each string or sentence entered by user should be a new line in text file save all the file in text file end display only those line which begins with capital letter.

```python
f=open("line.txt","a")
while True:
    s=input("Enter Line :")
    if s.upper()=="END":
        break
    else:
        f.write(s+"\n")
f.close()
f=open("line.txt")
d=f.readlines()
for i in d:
    if i[0].isupper():
        print(i)
```

# WAP to read line by line from a given files file1 and file2 and write into file3.

```python
f1=open("demo1.txt")
f2=open("demo2.txt")
f=open("demo3.txt","w")
s1=f1.readlines()
s2=f2.readlines()
l=0
if len(s1)>len(s2):
    l=len(s1)
else:
    l=len(s2)
for i in range(l):
    if len(s1)>i:
        f.write(s1[i])
    if len(s2)>i:
        f.write(s2[i])   
f.close()
```

# WAP to read a text file and print the text file where the word starting with 'i' it should be in reverse order.

```python
f=open("demo1.txt")
s=f.read().split()
for i in range(len(s)):
    l=s[i]
    if l[0]=="i" or l[0]=="I":
        print(l[::-1])
    else:
        print(l)
```
