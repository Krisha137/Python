```python
# class --> Student
# class Variable --> Branch
# Constructor --> RollNo , Name

# Methods :
# 1. enter_marks --> enter 5 Subject marks and store in List.
# 2. display --> rollNo,Name,Branch,Marks
# 3. calculater_avg --> calculate the avg marks & display
```

```python
class Student:
    Branch="CSE"
    def __init__(self,rollNo,Name):
        self.rollNo=rollNo
        self.name=Name
    def enter_marks(self):
        self.mark_list=[]
        for i in range(5):
            marks=int(input(f"Enter Marks Of Subject {i+1} :"))
            self.mark_list.append(marks)
    def display(self):
        print("-"*16)
        print("Student Name : ",self.name)
        print("Student RollNo : ",self.rollNo)
        print("Student Branch : ",self.Branch)
        if len(self.mark_list)>0:
            print("Student's Marks are :",self.mark_list)
    def calculater_avg(self):
        sum=0
        c=0
        for i in self.mark_list:
            sum+=i
            c+=1
        avg=sum/c
        print(self.name,"'s Average is :",avg)
```

```python
s=Student(11,"Krisha")
s.enter_marks()
s.display()
```

```python
s1=Student(12,"Janvi")
s1.enter_marks()
s1.display()
```

```python
s2=Student(13,"Priyal")
s2.enter_marks()
s2.display()
s2.calculater_avg()
```

# WAP to create class String and name methodes to count uppercase,lowercase,vovels,consonents,spaces and display thedetails of the that count.

```python
class String_count:
    def __init__(self,s):
        self.s=s
        self.upper=0
        self.lower=0
        self.space=0
        self.vovel=0
        self.conso=0
    def uppercase(self):
        for i in self.s:
            if i.isupper():
                self.upper+=1
    def lowercase(self):
        for i in self.s:
            if i.islower():
                self.lower+=1
    def spaces(self):
        for i in self.s:
            if i.isspace():
                self.space+=1
    def vovels(self):
        for i in self.s:
            if i.upper() in ["A","E","I","O","U"]:
                self.vovel+=1
    def consonents(self):
        for i in self.s:
            if i.isalpha():
                if i.upper() not in ["A","E","I","O","U"]:
                    self.conso+=1
    def display(self):
        self.uppercase()
        print("Upper Case :",self.upper)
        self.lowercase()
        print("Lower Case :",self.lower)
        self.spaces()
        print("Spaces :",self.space)
        self.vovels()
        print("Vovels :",self.vovel)
        self.consonents()
        print("Consonents : ",self.conso)
```

```python
s=String_count("Hello Sir.I am A Student.")
s.display()
```

```python
s=String_count(input("Enter String :"))
s.display()
```
