```python
# 1. lstrip() --> it will removing only left sides spaces
# 2. rstrip() --> it will removing only right sides spaces
# 3. strip()  --> it will removing both sides spaces
# mark not removing center spaces.
s="banana  "
print(len(s))
x=s.rstrip()
print(len(x))
print(x)
```

# Changing the case of string.

```python
# 1. upper() 
# 2. lower()
# 3. swapcase()  --> swap the case like upper in lower and lower in upper
s=" Hello World"
x=s.upper()
y=s.lower()
z=s.swapcase()
print(x)
print(y)
print(z)
```

# to check the type of characters present in a string(check function)
#### Answer only in true or false

```python
# 1. isalnum() : Returns True if all characters are alphanumaric(a-z,A-Z,0-9)
# eg.
x="company123"
print(x.isalnum())
y="company 123"
print(y.isalnum())
z="company1_23"
print(z.isalnum())
a="company_123"
print(a.isalnum())
```

```python
# 2. isalpha() : All the character's are alphabate
x="companyX"
print(x.isalpha())
y="company123"
print(y.isalpha())
```

```python
# 3. isdigit() : All the character's are digit
x="1234"
print(x.isdigit())
y="company123"
print(y.isdigit())
```

```python
# 4. isnumaric() : it is same as isdigit() and also include roman numbers
x="1234"
print(x.isdigit())
y="company123"
print(y.isdigit())
```

```python
# 5. islower()
t="hello world"
x=t.islower()
print(x)
```

# WAP print first,middle,last character of string if length is odd otherwise string.
