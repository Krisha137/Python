# WAP to append ,delete and display elements of a list using class.

```python
class List:
    def __init__(self):
        self.l=[]
    def l_append(self):
        a=input("Enter Data For append :")
        self.l.append(a)
    def l_delete(self):
        a=input("Enter Data For Delete :")
        self.l.remove(a)
    def display(self):
        for i in self.l:
            print(i,end=" ")
        print()
```

```python
l=List()
while True:
    print("-"*16)
    print("1. Append")
    print("2. Delete")
    print("3. Display")
    print("4. Exit")
    print("-"*16)
    n=int(input("Enter No : "))
    if n==1:
        l.l_append()
    elif n==2:
        l.l_delete()
    elif n==3:
        l.display()
    elif n==4:
        break
    else:
        print("Please Enter Valid No From (1,2,3) :")
```

# WAP to create class ATM and make methods to createpin,changepin,checkbalance and withdrow also menu.

```python
class ATM:
    def __init__(self):
        self.balance=100000
        self.pin=0
        self.menu()
    def createPin(self):
        while True:
            p=int(input("Enter Pin :"))
            if len(str(p))==4:
                self.pin=p
                print("Pin Created Successfully.")
                break
            else:
                print("Pin's Length must be 4.")
    def changePin(self):
        while True:
            p=int(input("Enter Old Pin :"))
            if p==self.pin:
                while True:
                    n=int(input("Enter New Pin :"))
                    if len(str(n))==4:
                        self.pin=n
                        print("Pin Created Successfully.")
                        break
                    else:
                        print("Pin's Length must be 4.")
                break
            else:
                print("Wrong Pin.")
    def checkBalance(self):
        while True:
            p=int(input("Enter Your Pin :"))
            if p==self.pin:
                print("üëãüèº------Hi Siri-----üëãüèº")
                print("Your Balance is ü§ëü§ë: ",self.balance)
                break
            else:
                print("Wrong Pin.")
    def withDrow(self):
        while True:
            p=int(input("Enter Your Pin :"))
            if p==self.pin:
                w=int(input("Enter Amount For Withdrow :"))
                if w<=self.balance:
                    self.balance=self.balance-w
                    print("üí∏üí∏üí∞----Withdrowed----üí∏üí∏üí∞")
                else:
                    print("‚ùå‚ùå---Sorry Siri---‚ùå‚ùå")
                    print("Your Balance is Low...")
                break
            else:
                print("Wrong Pin.")
    def menu(self):
        self.createPin()
        while True:
            print("-"*16)
            print("1. Change Pin")
            print("2. Check Balance")
            print("3. WithDrow")
            print("0. Exit")
            print("-"*16)
            n=int(input("Enter No : "))
            if n==0:
                break
            elif n==1:
                self.changePin()
            elif n==2:
                self.checkBalance()
            elif n==3:
                self.withDrow()
            else:
                print("Please Enter No From (0,1,2,3,4)")     
```

cheese

```python
class pizza:
    def __init__(self):
        self.price=0
        self.size={}
        self.toping1={}
        self.toping2={}
        self.cheesing={}
        self.menu()
    def pizza_size(self):
        s=input("Enter Pizza Size :")
        if s.lower()=="small":
            self.price+=50
            self.size[s.lower()]=50
        elif s.lower()=="medium":
            self.price+=100
            self.size[s.lower()]=100
        elif s.lower()=="large":
            self.price+=200
            self.size[s.lower()]=200
        else:
            print("Please Enter Valid Size.")
    def pizza_topping(self):
        print("Toping category 1 : corn,tomato,capsicum,onion")
        print("Toping category 2 : brocolli,Mashrom,olives")
        t=input("Enter Pizza Toping.")
        if t.lower() in ["corn","tomato","capsicum","onion"]:
            self.price+=20
            if t.lower() in self.toping1:
                self.toping1[t.lower()]+=1
            else:
                self.toping1[t.lower()]=1
        elif t.lower() in ["brocolli","Mashrom","olives"]:
            self.price+=50
            if t.lower() in self.toping2:
                self.toping2[t.lower()]+=1
            else:
                self.toping2[t.lower()]=1
        else:
            print("Sorry Your Topping Category is not available.")
    def pizza_cheesing(self):
        print("Available Cheese : cheese ,mozarella,feta,chedder")
        c=input("Enter Pizza Cheesing.")
        if c.lower() in ["cheese","mozarella","feta","chedder"]:
            self.price+=50
            if c.lower() in self.cheesing:
                self.cheesing[c.lower()]+=1
            else:
                self.cheesing[c.lower()]=1
        else:
            print("Sorry Your Cheesing Category is not available.")
    def pizza_price(self):
        c=int(input("How many Pizza You Want :"))
        self.price=self.price*c
        print("-----Hi Siri-----")
        print("Your Pizza Size : ")
        for i,j in self.size:
            print(i," - ",j)
        print("Your Pizza Topping Are : ")
        for i,j in self.toping1:
            print(i," - ",j)
        for i,j in self.toping2:
            print(i," - ",j)
        print("Your Pizza Cheesing Are :")
        for i,j in self.cheesing:
            print(i," - ",j)
        print("You Buy ",c," Pizza")
        print("Your Pizza'a Total Price Is : ",self.price)
        
    def menu(self):
        self.pizza_size()
        self.pizza_topping()
        self.pizza_cheesing()
        self.pizza_price()
```
