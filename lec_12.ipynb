# WAP to find position of the word.

```python
s="abcdabcacdabxyzabcabc"
x=s.find("abc")
print("Position : ",x)
for i in range(len(s)):
    z=s.find("abc",x+1)
    x=z
    if x==-1:
        break
    print("Position : ",z)
```

```python
s="abcdabcacdabxyzabcabc"
sub="abc"
pos=-1
flag=False
n=len(s)
while True:
    pos=s.find(sub,pos+1,n)
    if pos==-1:
        break
    print("Found at position : ",pos)
    flag=True
if flag==False:
    print("Not Found.")
```

# WAP to check the validity of the password.the primory condition of the password is given as below.
### 1. minimum 8 charactors.
### 2. The alphabets shold be between [a-z].
### 3. Atleast one uppercase between (A-Z).
### 4. Atleast onedigit between [0-9].
### 5. Atleast one charactor from [ _,@,$].

```python
p=input("Enter Password : ")
b=True
upper=0
lower=0
sc=0
digit=0
if len(p)>=8:
    for i in range(len(p)):
        if p[i].isupper():
            upper+=1
        elif p[i].islower():
            lower+=1
        elif p[i].isdigit():
            digit+=1
        elif p[i]=="@" or p[i]=="_" or p[i]=="$":
            sc+=1
        else:
            pass
    total=upper+lower+digit+sc
    if upper>=1 and lower>=1 and digit>=1 and sc>=1 and total==len(p):
        print("Password Is Valid.")
    else:
        print("Password is invalid.")
else:
    print("Password is invalid.")
```
