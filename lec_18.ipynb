## File Access Modes

```python
# --> Type of operations possible in opened file.
# --> how the file will be used once its opened.
# --> modes also define the location of the file handle in the file.
# --> File handle is like a cursor,which defines from where the data has to be read or writen in the file.
# --> Six Access modes (r,r+,w,w+,a,a+)
```

### 3. Write Only ( ' w ' )

```python
# --> Open text file for Writing.
# --> Handle is positioned at the begining of the file.
# --> for existing file,the data is truncated and overwritten.
# --> creates the file if file does not exist.
```

```python
f=open("demo.txt","w") # file will be blank.this is called truncated.
```

```python
f=open("demo.txt","w")
f.write("python-1\n") # output come when code run 2 times becase of close()
f.write("python-2\n")
f.write("python-3\n")
```

```python
f=open("demo.txt","w")
f.write("python-1\n") # output will come in 1 time run.
f.write("python-2\n")
f.write("python-3\n")
f.close()
```

```python
with open ("new.txt","w") as f:
    f.write("python-1\n") # it will create new file and write the data in 1 time here no need to write f.close()
    f.write("python-2\n")
    f.write("python-3\n")
```

```python
with open ("new.txt","w") as f:
    f.write("python-1\n") 
    f.write("python-2\n")
    f.write("python-3\n")
    print(f.closed) # out of indentation bydefault file  will close. no need to write f.close()
print(f.closed)
```

```python
f=open("demo.txt","r")
f.read()
print(f.closed)
f.close() # here we have to close the file manually.
print(f.closed)
```

```python
f=open("demo.txt","w")
print(f.read())
f.close()
```

### 5. append mode ( ' a ' )

```python
# --> Open text file for Writing.
# --> Handle is positioned at the end of the file.
# --> The data being written will be inserted at the end,after the existing data.
# --> creates the file if file does not exist.
```

```python
f=open("demo.txt","a")
f.write("New Data \n")
f.write("ljiet")
f.close()
```

```python
f=open("demo.txt","a")
f.write("New Data \n")
f.write("ljiet")
f.seek(0) # Data always write at end . bydefault
f.write("Xyz")
f.close()
```

# WAP to Copy file contant in new file.

```python
f=open("text.txt","r")
n=open("n.txt","w")
for i in f.read():
    n.write(i)
n.close()
```

```python
with open("demo.txt") as f1:
    a=f1.read()
with open("newf.txt","w") as f2:
    f2.write(a)
```

# WAP to Copy file contant in new file.But new Contant is in uppercase.

```python
f=open("text.txt","r")
n=open("ne.txt","w")
for i in f.read():
    n.write(i.upper())
n.close()
```

```python
with open("demo.txt") as f1:
    a=f1.read()
with open("newf2.txt","w") as f2:
    f2.write(a.upper())
```
