```python
%%writefile 12_matplotlib_basic.py
import streamlit as st
import matplotlib.pyplot as plt
import numpy as np

st.title("Matplotlib + Streamlit Demo")

x=np.arange(1,11)
y=np.random.randint(50,100,size=10)

#---------------
# Line Chart
# --------------
st.subheader("Line Chart (Matplotlib)")
plt.figure(figsize=(6,4))
plt.plot(x,y,marker='o')
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.title("Marks Of 10 Students")
st.pyplot(plt)

plt.clf() # clear the figure (stop overlaping)

# -------------
# Bar Chart 
# -------------
st.subheader("Bar Chart (Matplotlib)")
plt.figure(figsize=(6,4))
plt.bar(x,y)
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.title("Marks Bar Chart")

st.pyplot(plt)

# -----------------
# Histogram
# -----------------

st.subheader("Histogram")
plt.figure(figsize=(5,3))
plt.hist(y)
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.title("Marks Histogram")
st.pyplot(plt)

# -------------
# Scatter Plot
# -------------

st.subheader("Scatter plot")
plt.figure(figsize=(5,3))
plt.scatter(x,y,marker="o")
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.title("Scatter Plot")
st.pyplot(plt)

# python -m streamlit run 12_matplotlib_basic.py
```

# Example 13 : Matplotlib Streamlit lnbuilt Graphs 

```python
%%writefile 13_streamlit_graph.py
import streamlit as st
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

st.title("Matplotlib Streamlit Graph Demo")

marks=np.random.randint(50,100,size=10)
attendance=np.random.randint(50,100,size=10)

chart_data=pd.DataFrame({
    "Marks":marks,
    "Attendence":attendance
})

st.subheader("Line Chart")
st.line_chart(chart_data)

st.subheader("Area Chart")
st.area_chart(chart_data)

st.subheader("Bar Chart")
counts,bins=np.histogram(marks,bins=3)
st.bar_chart(counts)
```

# WAP to make Student Marks & Feddback Form

```python
%%writefile task_1.py
import streamlit as st
from datetime import date,time
import pandas as pd

st.title("📝Student Marks & Feddback Form 🖋")
st.header("1.Student Information")
col1,col2=st.columns([1,2])

with col1:
    no=st.text_input("Enrollment Number")
    name=st.text_input("Student Name")
with col2:
    sem=st.selectbox("Semester",["1","2","3","4","5","6","7","8"])
    div=st.text_input("Division")
    
examDate=st.date_input("Exam Date ",value=date.today())

st.header("2.Marks Entry")
python=st.number_input("Python-1 Marks(out of 100)",min_value=0,max_value=100)
fsd=st.number_input("FSD-1 Marks(out of 100)",min_value=0,max_value=100)
de=st.number_input("DE Marks(out of 100)",min_value=0,max_value=100)

st.header("3.Feedback")
feed=st.slider("How well did you understand the subject?",min_value=0,max_value=10,value=7)
part=st.radio("Class Participation",["Low","Medium","High"])
text=st.text_area("Additional Comments")
if st.button("Submit Record"):
    st.success("Record Submited Successfully.")
    df=pd.DataFrame({
        'Enrollment No ':[no],
        'Name':[name],
        'Semester':[sem],
        'Division':[div],
        'ExamDate':[examDate],
        'Python Marks':[python],
        'FSD Marks':[fsd],
        'DE Marks':[de],
        'Feedback':[feed],
        'Class Participation':[part],
        'Comments':[text]
    })
    st.write("Generated Data")
    st.dataframe(df)
    csv=df.to_csv(index=False).encode('utf-8')
    st.download_button(label='Download as CSV',data=csv,file_name='sample.csv',mime="text/csv")
```

# WAP to create a bas plot a cources no.of students enroll in this course using dictionary with appropeate axis x and y and color of bar should be green.

```python
import matplotlib.pyplot as plt

data={"C":20,"C++":11,"JAVA":15,"Python":30,"FSD":45,"JavaScript":10,"PHP":21,"R":14}
x=[]
y=[]
for i in data:
    x.append(i)
    y.append(data[i])
plt.bar(x,y,color="m")
plt.xlabel("Course",c="blue")
plt.ylabel("Marks",c="blue")
plt.title("Marks Bar Chart",c="r")
plt.show()
```
