# WAP to count no. of lines ,words, characters and spaces from the given text file.

```python
with open("demo.txt") as f:
    b=f.read()
    x=b.split()
    print("Words : ",len(x))    
    f.seek(0)
    print("Lines : ",len(f.readlines()))
    y=len(b)-b.count(" ")-b.count("\n")
    print("Characters : ",y)
    sp=b.count(" ")
    print("Space : ",sp)
```

# WAP your solution should promt for a file name and display the text file 5 lines at a time posing each time to ask the user to press any key to continue and press q to quit.

```python
name=input("Enter File Name : ")
fn=name+".txt"
text_lined=open(fn)
num_line=5
for index,line in enumerate(text_lined,1):
    print(index,":",line)
    if index%num_line==0:
        x=input("Enter any key to continue or press q to quite")
        if x.lower()=="q":
            break
        else:
            continue
```

# WAP to compare two files.

```python
f1=open("demo1.txt")
f2=open("demo2.txt")
r1=f1.read()
r2=f2.read()
c=n=1
for k in range(len(r1)):
    if r1[k]==r2[k]:
        c+=1
        if r1[k]=="\n":
            n+=1
            c=1
    else:
        print("Line ",n,"Position ",c)
        c+=1
```

# WAP to remove commant and add code in text file.

```python
f=open("demo1.txt")
d=open("demo2.txt","w")
while True:
    b=f.readline()
    if len(b)!=0:
        if b[0]=="#":
            continue
        else:
            if "#" in b:
                for index in range(1,len(b)):
                    if b[index]=="#":
                        d.write(b[0:index]+"\n")
            else:
                d.write(b)
    else:
        break
f.close()
d.close()
```
