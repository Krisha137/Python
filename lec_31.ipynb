# Bar Chart 

### bar()

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10]) # y's cordinate that will work as a bar
plt.bar(x,y) # It will give Vertical bar
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10]) 
plt.barh(x,y) # It will give Horizontal bar
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10]) 
plt.bar(x,y,color="pink") # color
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10]) 
plt.bar(x,y,color="pink",width=0.4) # Width --> Bydefult is 0.8 --> for vertical
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10]) 
plt.barh(x,y,color="pink",width=0.4) 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10]) 
plt.barh(x,y,color="pink",height=0.4) # height --> Bydefult is 0.8 --> for horizontal
plt.show()
```

# Histogram

```python
# --> Histogram is a graph showing frequency distribution.
# --> It shows the no. of observations within each given interval.
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,1,2,3,3,3,3,4,4,5,7,8,8,9,10,10]) # It will give ans acording it's frequency
plt.hist(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,1,2,3,3,3,3,4,4,5,7,8,8,9,10,10]) 
plt.hist(x,bins=5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,1,2,3,3,3,3,4,4,5,7,8,8,9,10,10]) 
plt.hist(x,bins=5,orientation="vertical") # By defult orientation is vertical
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,1,2,3,3,3,3,4,4,5,7,8,8,9,10,10]) 
plt.hist(x,bins=5,orientation="horizontal")
plt.show()
```

# Area Fill Between Two Lines

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color="green")
plt.plot(time,expense,color="red")
```

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color="green")
plt.plot(time,expense,color="red")
plt.fill_between(time,income,expense,where=(income>expense),color="green",alpha=0.25,label="positive")
plt.fill_between(time,income,expense,where=(income<=expense),color="red",alpha=0.25,label="negative")
```

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color="green")
plt.plot(time,expense,color="red")
plt.fill_between(time,income,expense,where=(income>expense),color="green",alpha=0.25,label="positive",interpolate=True)
plt.fill_between(time,income,expense,where=(income<=expense),color="red",alpha=0.25,label="negative",interpolate=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color="green")
plt.plot(time,expense,color="red")
plt.fill_between(time,income,expense,where=(income>expense),color="green",alpha=0.25,label="positive",interpolate=True)
plt.fill_between(time,income,expense,where=(income<=expense),color="red",alpha=0.25,label="negative",interpolate=True)
plt.legend()
```
