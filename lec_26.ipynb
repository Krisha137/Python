## OverLoading

```python
# --> Allow Programers to write methods that can process a variety of different types of functioalities eith the same name.
# --> Method overloading is a compile type Polymorphism using which we can define two or more methods in same class with 
#     the same name but with a different parameter list.
```

```python
def sum_num(*args):  # It's not actual concept of overloading but this is similer to overloading.
    result=0
    for num in args:
        result+=num
    print("Sum :",result)
print("Similer Method Overloading Consept")
print("Single Argument --> ",end="")
sum_num(10)
print("Two Argument --> ",end="")
sum_num(10,20)
print("Multiple Argument --> ",end="")
sum_num(10,20,30,40,50)
```

## Method OverRiding

```python
# --> A child class method overrides the parent class method of the same , name, parameters and return type. It is used to
#     overwrite or redefine a parent class method in the derived class.
```

```python
class A:
    def first(self):
        print("First Function Of class A")
    def second(self):
        print("Second Function Of class A")
class B(A):
    def first(self): # this is called overidden Method
        print("Redefined function Of class A in class B")
    def display(self):
        print("Display Function Of Child class")
```

```python
obj=B()
obj.first()
obj.second()
obj.display()
```

```python
obj=A()
obj.first()
obj.second()
obj.display()
```

```python
A().first()
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __add__(self,other):
        return self.a+other.a
```

```python
obj1=A(5)
obj2=A(3)
obj3=A("Hello")
obj4=A("Hi")
```

```python
print(obj1+obj2) # + call bydefult __add__
print(obj3+obj4) # when it call by operator then write like this
```

```python
print(obj1.__add__(obj2))
print(obj3.__add__(obj4))
```

```python
print(A.__add__(obj1,obj2))
print(A.__add__(obj3,obj4))
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __gt__(self,other):
        if self.a>other.a:
            return True
        else:
            return False
```

```python
obj1=A(5)
obj2=A(7)
if obj1>obj2: # > it's call __gt__
    print("obj1 is greater") 
else:
    print("obj2 is greater")
```

```python
obj1=A(8)
obj2=A(7)
if obj1>obj2:
    print("obj1 is greater") 
else:
    print("obj2 is greater")
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __lt__(self,other):
        if self.a<other.a:
            return True
        else:
            return False
```

```python
obj1=A(8)
obj2=A(7)
if obj1<obj2:
    print("obj1 is Lesser") 
else:
    print("obj2 is Lesser")
```

# Inheritance

```python
# --> We can use existing class to create a new class rather than creating it from a scratch.
# --> Allows the program to reuse the code.
# --> With inheritance,child class gains access to all data members , functions and properties defined in the parent class.
# --> A child class may also offwe its particular implementation of the parent class's functions .
# --> A derived class can inherit from its base class by simple putting its name in brackets after the derived class name.
```

### 1.single Inheritance(Simple Inheritance)

```python
class parent:
    def fun1(self):
        print("Hello Parent")
class child(parent):
    def fun2(self):
        print("Hello Child")
```

```python
obj=child()
obj.fun1()
obj.fun2()
```

### 2. Multiple Inheritance

```python
class parent1:
    def fun1(self):
        print("Hello Parent 1")
class parent2:
    def fun2(self):
        print("Hello Parent 2")
class child(parent1,parent2):
    def fun3(self):
        print("Hello Child")
```

```python
obj=child()
obj.fun1()
obj.fun2()
obj.fun3()
```

### 3. Multilevel Inheritance

```python
class Grandparent:
    def fun1(self):
        print("Hello Grandparent")
class Parent(Grandparent):
    def fun2(self):
        print("Hello Parent")
class Child(Parent):
    def fun3(self):
        print("Hello Child")
```

```python
obj=Child()
obj.fun1()
obj.fun2()
obj.fun3()
```

### 4. Hierachical Inheritance

```python
class parent:
    def fun1(self):
        print("Hello Parent")
class Child1(parent):
    def fun2(self):
        print("Hello Child 1")
class Child2(parent):
    def fun3(self):
        print("Hello Child 3")
```

```python
obj=Child1()
obj.fun1()
obj.fun2()
obj.fun3()
```

```python
obj=Child2()
obj.fun1()
obj.fun3()
obj.fun2()
```

### 5. Hybrid

```python
class parent1:
    def fun1(self):
        print("Hello Parent 1")
class parent2:
    def fun2(self):
        print("Hello Parent 2")
class child1(parent1):
    def fun3(self):
        print("Hello Child 1")
class child2(child1,parent2):
    def fun4(self):
        print("Hello Child 2")
```

```python
obj1=child1()
obj2=child2()
obj1.fun1()
obj1.fun3()
obj2.fun1()
obj2.fun2()
obj2.fun3()
obj2.fun4()
```
