# Important Function of set

```python
# 1. len()

s=set("Apple")
print(len(s))
```

```python
# 2. add(x) :
#            --> add item x to the set
#            --> if you want to add 1 then use it.

s={10,20,30}
s.add(40)
print(s)
```

```python
# 3. update() :
#             --> update(x,y,z)
#             --> if you want to add multiple value then use it.

s={10,20,30}
l=[40,50,60,10]
s.update(l,range(5))
print(s)
```

```python
# 4. copy() :
#           --> for create clone.

s={10,20,30}
l=s.copy()
print(s)
print(l)
l.add(40)
print(s)
print(l)
```

```python
# 5. pop() :
#          --> Remove random element.removed element is not fix.

s={10,20,30}
print(s.pop())
print(s)
```

```python
x={"a","b","c","d"}
print(x.pop())
print(x)
```

```python
# 6. remove() :
#             --> Removed element which is given in parameter.if element not present in the set then it will gave error.

s={10,20,30,40}
s.remove(10)
print(s)
s.remove(50)
print(s)
```

```python
s={10,20,10,30,40}
s.remove(10)
print(s)
s.remove(50)
print(s)
```

```python
# 7. discard() :
#              --> same as remove.but if element not present in the set then it will not gave error.

s={10,20,30,40}
s.discard(10)
print(s)
s.discard(50)
print(s)
```

```python
# 8. clear() :
#            --> It clear the set.

s={10,20,30}
s.clear() 
print(s)   # set()
```

# Lambda Function

```python
# --> function without name. 
# --> nameless function is called anonymous function or lambda function.
```

```python
def square(n):
    return n*n
print(square(5))
print(square(7))
```

```python
s=lambda n:n*n
#        ^  ^
#        ^  it will return
#  it will pass in the parameter.

print(s(5))
print(s(7))
```

```python
s=lambda a,b:a+b
print(s(10,20))
```

```python
s=lambda a,b:(a+b,a-b,a*b,a/b)  # ans in tuple
print(s(10,20))
```

```python
s=lambda x:x.upper()
print(s("Python"))
```

```python
words=["apple","banana","cherry","kiwi"]
s=sorted(words,key = lambda x:len(x)) # ans based on length
print(s)
```

```python
words=["apple","banana","cherry","kiwi"]
s=sorted(words,key = lambda x:len(x),reverse=True) # for reverse
print(s)
```

```python
a=[(1,3),(4,2),(5,8),(9,1)]
a.sort(key=lambda x:x[1]) # sort by second element
print(a)
```

```python
a=[(1,3),(4,2),(5,8),(9,1)]
s=sorted(a,key=lambda x:x[1]) # sorted use for all like list,tuple.....
print(s)
```

```python
# map() , reduce() , filter() :
#                             --> by which we can use lambda function.
```

```python
# 1. map() :
#           --> for every element present in given sequence, apply some functionality and generate new element 
#               with required modification.
#         Syntex :
#                 map(function,sequence)
```

```python
# by userdefine function

l=[1,2,3,4,5]
def double(x):
    return 2*x
l1=list(map(double,l))
print(l1)
```

```python
# by lambda function

l=[1,2,3,4,5]
l1=list(map(lambda x:x*2,l))
print(l1)
```

```python
l1=[1,2,3,4,5]
l2=[2,3,4,5,6]
l3=list(map(lambda x,y:x*y,l1,l2)) # element wise multiplication (l1,l2 => l3)
print(l3)
```

```python
# 2. reduce() :
#              --> reduces sequence of elements into a single element by applying specific function.
#       Syntax :
#               reduce(function,sequence)
#              --> reduce() function is present in functools module so we need to import the module before using 
#                  this function.
```

```python
from functools import *
# import functools --> then you have to write like this functools.reduce() when it's use

l=[10,20,30,40,50]
sum=reduce(lambda x,y:x+y,l)
print(sum)
mul=reduce(lambda x,y:x*y,l)
print(mul)
```

```python
# 3. filter() :
#              --> used to filter values from given sequence based on some condition.
#              --> filter(function,sequence)
#              --> function argument is responsible to perform conditional check on sequence.
```

```python
# by userdefine function

def even(x):
    if x%2==0:
        return True
    else:
        return False
l=[0,5,10,15,20,25,30]
l1=list(filter(even,l))
print(l1)
```

```python
# by lambda function

l=[0,5,10,15,20,25,30]
l1=list(filter(lambda x:x%2==0,l))
print(l1)
```

```python
check=lambda x:"Even" if x%2==0 else "Odd"
print(check(5))
print(check(8))
```

# WAP to calculate the sum of positive and negative number from the given list of numbers using lambda function.

```python
s=[2,4,-6,-9,11,-12,14,-5,17]
sum_p=lambda x:sum(num for num in x if num>0)
sum_n=lambda x:sum(num for num in x if num<0)
print(sum_p(s))
print(sum_n(s))
```

```python
l=[2,4,-6,-9,11,-12,14,-5,17]
p=sum(filter(lambda x:x>0,l))
n=sum(filter(lambda x:x<0,l))
print(p)
print(n)
```

# Input : B4A1D3 ; Output : ABD134

```python
s=input("Enter String with digits : ")
a=""
d=""
for i in s:
    if i.isalpha():
        a+=i
    elif i.isdigit():
        d+=i
l=sorted(a)+sorted(d)
for i in l:
    print(i,end="")
```
