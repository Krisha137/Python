```python
class tejas(Exception):
    def __init__(self,arg):
        self.msg=arg
```

```python
def add(a,b):
    if type(a)==str:
        raise tejas("Personalized msg")
    if type(b)==str:
        raise TypeError("b should be int.")
add("abc",'def')
```

```python
def checkbalanced():
    money=10000
    withdrow=7000
    try:
        balance=money-withdrow
        if balance<=2000:
            raise tejas("Low Balance.")
        print(balance)
    except tejas as e:
        print(e)
checkbalanced()
```

```python
class Employee:
    def __init__(self,name,age,salary): # self is bydefult parameter
        self.name=name
        self.age=age
        self.salary=salary
    def display(self):
        print("Employee Name : ",self.name)
        print("Age : ",self.age)
        print("Salary : ",self.salary)
```

```python
emp1=Employee("Arman",30,50000)
emp2=Employee("Dhairya",25,40000)
emp1.display()
emp2.display()
```

```python
class Employee:
    def __init__(self,name,age,salary): # self is bydefult parameter
        self.name=name
        self.age=age
        self.salary=salary
        print("Address Of Self : ",id(self))
    def display(self):
        print("Employee Name : ",self.name)
        print("Age : ",self.age)
        print("Salary : ",self.salary)
```

```python
emp1=Employee("Arman",30,50000)
emp2=Employee("Dhairya",25,40000)
emp1.display()
print("Address Of Object : ",id(emp1))
emp2.display()
print("Address Of Object : ",id(emp2))
print(emp1.name)
```
