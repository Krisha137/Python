# WAP to transpose matrix.

```python
l=[[1,2,3],[4,5,6],[7,8,9]]
print("Original Matrix : ")
print("-"*50)
for i in range(len(l)):
    for j in range(len(l[i])):
        print(l[i][j],end="")
    print()
print("-"*50)
print("Transpose Matrix : ")
print("-"*50)
for i in range(len(l)):
    for j in range(len(l[i])):
        print(l[j][i],end="")
    print()
```

# Dictionary 

```python
# --> represent group of object as key value pairs
# eg.
#    roll no - name
#    phone no - address
# --> duplicate keys not allowed but values can be duplicates.
# --> heterogeneous objects allowed for both key & values.
# --> Insertion order not preserved.
# --> indexing & slicing concept are not applicabe.
```

```python
# if you enter two diffrent values of same keys , then it update the value.

d={10:"Lucky",20:"Arman",30:"Dhairya",10:"Aryan"}
print(d)
```

# Creation of dictionary

```python
# d={} 
#  or
# d=dict()
```

```python
d=dict()
d[10]="Lucky"
d[20]="Arman"
d[30]="Dhairya"
print(d)
```

```python
d={}
d[10]="Lucky"
d[20]="Arman"
d[30]="Dhairya"
print(d)
```

# Accessing the elements of dictionary

```python
d={10:"Lucky",20:"Arman",30:"Dhairya",10:"Aryan"}
print(d[10])
print(d[20])
print(d[40]) # key error
```

# Updating elements of  dictionary

```python
d={10:"Lucky",20:"Arman",30:"Dhairya",10:"Aryan"}
d[40]="Smit"
print(d)
d[10]="Nayan"
print(d)
```

# Deleting the element of dictionary

```python
d={10:"Lucky",20:"Arman",30:"Dhairya",10:"Aryan"}
del d[10]
print(d)
del d[40]
print(d) # key error
```

```python
# delete dictionary
d={10:"Lucky",20:"Arman",30:"Dhairya",10:"Aryan"}
del d
print(d) # NameError
```

```python
# only data will be delete
d={10:"Lucky",20:"Arman",30:"Dhairya",10:"Aryan"}
d.clear()
print(d)
```
