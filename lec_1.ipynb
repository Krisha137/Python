# Identifier

- A name in Python Program is called Identifiers.
- class name,function name,module name,variable name.

eg. 

   Q=10     
   def test()    
   class Car

# Rules 

1.the only allowed characters in python are :
  - alphabet Symbols(a-z ,A-Z)
  - digits(0-9)
  - underscore symbol(-)
 

2.Identifier should not starts with digits.

```python
total123=100
```

```python
123total=100
```

3.Identifier are **Case Sensitive**

```python
total = 100
```

```python
Total=100
```

```python
TOTAL=100
```

4.we can not use reserved keywords as identifiers.

```python
def=10
```

```python
DEF=100
```

```python
if=10
```

```python
If=10
```

5. There is no Length limit For Python Identifier . But Not Recommended to use too lengthy Identifiers.

```python
a=10
```

```python
XYZ_ABC_PQR=10
```

6. If Identifiers starts eith Underscore( _ ) then it is **private identifier**.

```python
_a=10
```

7. If Identifiers starts with double underscore( _ _ ) then it is **strongly private identifier**

```python
__a=20
```

8. If identifier starts and ends with two underscore symbols then the identifier is **language defined special name** , which is also known as **magic methods**.

```python
__add__=10
```
