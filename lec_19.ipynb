# WAP to read the content of file 1 and print in reverse in file 2.

```python
f=open("demo.txt")
l=f.readlines()[::-1]
print(l)
f2=open("demo1.txt","w")
for i in range(len(l)):
    if i==0:
        f2.write(l[i]+'\n')
    else:
        f2.write(l[i])
f2.close()
f.close()
```

# WAP to take a line from file and make list of words from that line.

```python
f=open("demo.txt")
l=f.readlines()
n=[]
for i in range(len(l)):
    n.append(l[i].split())  # by default split consider space.
for i in range(len(n)):
    print(n[i])
```

# WAP to write the dictionary in file as a data.

```python
details={"name":"Alice","Age":"25","Degree":"CSE","Univercity":"LJ"}
d=open("details.txt","w")
for i,j in details.items():
    d.write(i)
    d.write(" : ")
    d.write(j)
    d.write("\n")
```

```python
details={"name":"Alice","Age":"25","Degree":"CSE","Univercity":"LJ"}
d=open("details.txt","w")
for i,j in details.items():
    l=i+" : "+j+"\n"
    d.write(l)
```

```python
# it considerd int also.

details={"name":"Alice","Age":25,"Degree":"CSE","Univercity":"LJ"}
with open ("myfile.txt","w") as f:
    for key,value in details.items():
        f.write(str(key)+" : "+str(value)+"\n")
```

# WAP to make file that conatins odd lines from another files's data.

```python
f=open("demo.txt")
l=f.readlines()
f2=open("Oddlines.txt","w")
for i in range(len(l)):
    if i%2==0:
        f2.write(l[i]+"\n")
```

# WAP to replace the word by new word.

```python
f=open("demo.txt")
l=f.readlines()
print(l)
f=open("demo.txt","w")
for i in range(len(l)):
    n=l[i].split()
    for j in n:
        if j=="Apple":
            j="Chery"
    print(n)
```
