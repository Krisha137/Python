# Make user define module which return ,if given data is string then it will give string's method and if given data is list then it will give list's method.

```python
%%writefile str_list.py
def my_string(s):
    print("Len : ",len(s))
    print("Upper : ",s.upper())
    print("Lower : ",s.lower())
    print("Title : ",s.title())
    print("Capitalize : ",s.capitalize())
    print("Is Alpha : ",s.isalpha())
    print("Is Alpanumaric : ",s.isalnum())
    print("Is Title : ",s.istitle())
    print("Is Digit : ",s.isdigit())
def my_list(l):
    print("Len : ",len(l))
    l.append(10)
    print("Append : ",l)
    l.extend([2,6,7])
    print("Extend : ",l)
    l.sort()
    print("Sort : ",l)
    l.reverse()
    print("Reverse : ",l)
```

```python
import str_list as l
s=eval(input("Enter Data : "))
if type(s)==str:
    l.my_string(s)
elif type(s)==list:
    l.my_list(s)
else:
    print("Given Data is not string or list...")
```

# Unit 8 : OOP & Exception Handling                                          

```python
# try --> block lets you test a block of code for errors.
# except --> block lets you handle the error.
# else --> lets you execute code when there is no error.
# finally --> lets you execute code,regardless of the result of try and except block.
```

```python
#       error
#       -----------> except ----
#       |                      |
# try ---                      |
#       |                      |
#       -----------> else  -----
#       no error               |
#                              ---->> finally
```

```python
def div(a,b):
    return a/b
print(div(5,0))
```

```python
def div(a,b):
    return a/b
try:
    print(div(5,0))
except:
    print("Exception Occured.")
else:
    print("No exception")
```

```python
def div(a,b):
    return a/b
try:
    print(div(5,2))
except:
    print("Exception Occured.")
else:
    print("No exception")
```

```python
def div(a,b):
    return a/b
try:
    print(div(5,0))
except:
    print("Exception Occured.")
else:
    print("No exception")
finally:
    print("Always print.")
```
