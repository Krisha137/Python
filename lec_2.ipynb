# # Global variable VS Local variable

```python
a="python" #global
def test():
    a="java" #local
    print(a)
test()
print(a)
```

```python
a="python"
def test():
    a="java"
    print(a)
print(a)
test()
```

```python
a="python"
def test():
    global a
    a="java"
    print(a)
print(a)
test()
```

```python
a="python"
def test():
    global a
    a="java"
    print(a)
test()
print(a)
```

  - --> Substraction
  * --> Multiplication  (str -->String Multiplication)
  / --> division
  % --> Modulus
  // --> Floor Division
  ** --> Exponent or power

```python
a=int(input("Enter num1 : "))
b=int(input("Enter num2 : "))
print("Addition : " ,a+b)
print("Substraction : ",a-b)
print("Multiplication : ",a*b)
print("Division : ",a/b)
print("Modulus : ",a%b)
print("Floor Division : ",a//b)
print("Exponent or power : ",a**b)
```
